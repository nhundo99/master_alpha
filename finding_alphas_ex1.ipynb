```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def alpha(data):
    close = data['Close']
    close_5_ago = data['Close'].shift(periods=5)
    
    alpha = -(close - close_5_ago)/close_5_ago
    return alpha
```

```python
data_aapl = pd.read_csv('../data/hist/AAPL.csv', parse_dates=['Date'])
data_msft = pd.read_csv('../data/hist/MSFT.csv', parse_dates=['Date'])
data_amzn = pd.read_csv('../data/hist/AMZN.csv', parse_dates=['Date'])
data_goog = pd.read_csv('../data/hist/GOOG.csv', parse_dates=['Date'])
data_meta = pd.read_csv('../data/hist/META.csv', parse_dates=['Date'])
```
